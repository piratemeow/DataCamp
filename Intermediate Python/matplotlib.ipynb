```python
# Matplotlib is a plotting library for the Python programming language and its numerical mathematics extension NumPy.
# It is used to visualize data in different graphs such as bar chart, histogram, scatter plot, pie chart, line plot, etc.
```

```python
import matplotlib.pyplot as plt # Importing mathplotlib.pyplot library as plt
```

```python
year = [1950, 1970, 1990, 2010] # Creating a list of years
population = [2.519, 3.692, 5.263, 6.972] # Creating a list of population in billions

plt.plot(year,population) # This is used to polt the graph. As how and where to plot the graph
# plot() method is used to create the line graph.
plt.show() # Displaying the graph. It can be omitted in jupyter notebook and some other IDEs but must be used in a python script

# if we want to see a scatter graph 
plt.scatter (year,population) # This is used to polt the graph. As how and where to plot the graph
# scatter() method is used to create the scatter graph.
plt.show() # Displaying the graph
```

```python
# We can change the scaling of the x and y axis

plt.scatter (year,population) # This is used to polt the graph. As how and where to plot the graph
plt.xscale('log') # Changing the scaling of x axis
plt.show() # Displaying the graph
```

```python
print(len(year),len(population)) # Printing the length of the list year
plt.scatter (year,population) # This is used to polt the graph. As how and where to plot the graph
plt.xlabel('Year') # Adding the label to x axis
plt.ylabel('Population') # Adding the label to y axis
plt.title('Population vs Year') # Adding the title to the graph
plt.yticks([0,2,4,6,8,10],['0','2B','4B','6B','8B','10B']) # Changing the y axis ticks. The first row represents the actual values that y axis can have
                                                           # The second row represents the values that will be displayed on the y axis
plt.show() # Displaying the graph
```

```python
# Histogram is a graphical representation of the distribution of numerical data.
# it is an estimate of the probability distribution of a continuous variable (quantitative variable)
# and was first introduced by Karl Pearson.
# Means it will show the frequencyt of the data
pop = [1,1,1,2,2,3,3,3,3,3,4,5,6,6,6,7,8,9,9,9,9,9,10,11,12]
print(len(pop))
plt.hist(pop,4) # This is used to polt the graph. As how and where to plot the graph. The bin number is set to 10 by default
# The second paprameter in his function is bin number and it is used to divide the data into bins. it is optional.
plt.xticks([1,3,6,9,11,12]) # Changing the x axis ticks. The first row represents the actual values that x axis can have
plt.show() # Displaying the graph
```

```python
col = ['red','green','blue','violet']
siz = [10,90,150,170]
plt.scatter(year,population,s=siz,c=col) # s parameter sets the size for the dots in the graph. For individual dot we can set different size by 
                                    # passing a list or for all the dots to be of same size we can pass a number
                                    # c parameter sets the color for the dots in the graph. For individual dot we can set different color by
                                    # passing a list or for all the dots to be of same color we can pass a number/color
```

```python
plt.scatter(year,population,s=100,c='red')
plt.text(1950,2.519,'First point') # This will add the text to the graph at the given coordinates
plt.grid(True) # This will add the grid to the graph
```
