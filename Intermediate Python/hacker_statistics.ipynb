```python
#numpy has a random number generator subpackage.
import numpy as np

# Seed the random number generator. It is important because the random number gets generated on the basis of the seed.
np.random.seed(123)
x= np.random.rand() # Generate a random number between 0 and 1. It is a float number.
print(x)
x = np.random.randint(0,2) # Generate a random number between 0 and 1. It is a integer number.
print(x)
```

```python
# Generating random walks
# A walk is a succession of steps determined by a random process. But the next step depends on the previous step.

outcome=[]
np.random.seed(123)
for x in range(10):
    coin = np.random.randint(0,2)
    if coin == 0:
        outcome.append("heads")
    else:
        outcome.append("tails")
print(outcome)

# Random walk
tails=[0]
np.random.seed(123)
for x in range(10):
    coin = np.random.randint(0,2)
    tails.append(tails[x]+coin)
print(tails)
```

```python
import matplotlib.pyplot as plt

# Plot the random walk

random_walk = []
np.random.seed(123)
for i in range(100000):
    tails=[0]
    for x in range(10):
        coin = np.random.randint(0,2)
        tails.append(tails[-1]+coin)
    random_walk.append(tails[-1])
print(random_walk)
plt.hist(random_walk, bins=10)
plt.grid(True)
plt.show()
```
